```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# plots to be rendered inside the notebook
%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./fuel_econ.csv')
print(fuel_econ.shape)
fuel_econ.head()
```

```python
plt.scatter(data = fuel_econ, x= 'displ', y='comb');
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

negative correlation between engine size and fuel efficiency there is a large amount of variability for the smaller displacement compare to larger ones

```python
# we can use seaborn as well
sb.regplot(data = fuel_econ, x= 'displ', y='comb'); # fit_reg = False will remove regression line
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
# example of log tans which is not useful here
def log_trans(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

sb.regplot(data = fuel_econ, x= 'displ', y='comb'); # fit_reg = False will remove regression line
tick_locs = [10, 20, 50, 100, 200, 500]
plt.yticks(log_trans(tick_locs), tick_locs);
```

```python
# combined fuel efficiency against year
sb.regplot(data = fuel_econ, x= 'year', y='comb'); # fit_reg = False will remove regression line
```

```python
sb.regplot(data = fuel_econ, x= 'year', y='comb', x_jitter = 0.3, scatter_kws = {'alpha': 1/5});
# the intresting areas in the center is still pretty indistinguishable
```

```python
sb.regplot(data = fuel_econ, x= 'year', y='comb', x_jitter = 0.3, scatter_kws = {'alpha': 1/20});
```

**Task 1**: Let's look at the relationship between fuel mileage ratings for city vs. highway driving, as stored in the 'city' and 'highway' variables (in miles per gallon, or mpg). Use a _scatter plot_ to depict the data. What is the general relationship between these variables? Are there any points that appear unusual against these trends?

```python
sb.regplot(data = fuel_econ, x= 'city', y='highway', x_jitter = 0.3, scatter_kws = {'alpha': 1/5}, fit_reg = False);
plt.xlabel('City Fuel Eff. (mpg)')
plt.ylabel('Highway Fuel Eff. (mpg)');
```

```python
plt.scatter(data = fuel_econ, x = 'city', y = 'highway', alpha = 1/8)
# plt.plot([10,60], [10,60]) # diagonal line from (10,10) to (60,60)
plt.xlabel('City Fuel Eff. (mpg)')
plt.ylabel('Highway Fuel Eff. (mpg)')
```
